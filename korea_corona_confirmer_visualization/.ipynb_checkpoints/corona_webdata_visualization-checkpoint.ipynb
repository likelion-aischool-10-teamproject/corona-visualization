```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
from selenium.webdriver import ActionChains
import pandas as pd
import os
import folium
import json
```

```python
# 셀레니움 이용 코로나 현황 페이지 데이터 수집
driver = webdriver.Chrome('../chromedriver_90')
url = 'https://coronaboard.kr/'
driver.get(url)
time.sleep(3)  # 홈페이지 로딩 시간 3초
```

```python
# 대한민국 현황 셀렉트박스 클릭
sel_box = driver.find_element_by_xpath('//*[@id="korea-slide"]/div/div[4]/div/button/div/div/div')
ActionChains(driver).move_to_element(sel_box).perform()
time.sleep(1)

sel_box.click()
time.sleep(1)
```

```python
# 인구수 체크 추가
sel_option = driver.find_element_by_xpath('//*[@id="bs-select-2-4"]')
sel_option.click()
time.sleep(1)

sel_box.click()
time.sleep(1)
```

```python
# 대한민국 현황 테이블에서 데이터 수집
all_data = []

for i in range(2, 9) :
    temp_xpath = f'//*[@id="kr-table"]/div/div/table/tbody/tr/td[{i}]'
    sel_ele = driver.find_elements_by_xpath(temp_xpath)
    column_data = []

    for one in sel_ele :
        column_data.append(one.text)
        
    all_data.append(column_data)
    print(column_data)
    print()
```

```python
# 수집된 데이터 이용하여 데이터프레임 생성 및 파일 저장
data_df_ori = pd.DataFrame({'지역' : all_data[0],
                            '확진자' : all_data[1],
                            '치료중' : all_data[2],
                            '사망자' : all_data[3],
                            '격리해제' : all_data[4],
                            '발생률' : all_data[5],
                            '인구수' : all_data[6]})

data_df_ori.head()

# # 연월일 문자
day = time.strftime('%Y-%m-%d', time.localtime())

file_name = f'{day}_국내_코로나_현황.xlsx'
data_df_ori.to_excel(file_name, index=False)

if file_name in os.listdir() :
    print(f'{file_name} 저장 완료')
else : 
    print(f'{file_name} 저장 실패')
```

```python
# 셀레니움 종료
driver.close()
```

```python
# 원본데이터 복사
data_df = data_df_ori.copy()

data_df.head()
```

```python
# 확진자 데이터 1일 확진자, 합계 확진자로 분류
# 지역이 '검역'인 행 삭제
data_df['확진자_합계'] = data_df['확진자'].str.split('\n').str[0]
data_df['확진자1일'] = data_df['확진자'].str.split('\n').str[1]
data_df = data_df.drop(data_df[data_df['지역'] == '검역'].index)

data_df.head()
```

```python
# folim 이용 기본 지도생성 및 확인
map = folium.Map(location=[36, 127.51], zoom_start=6, tiles='cartodbpositron') 
map
```

```python
# 전국 시도별 geojson 파일
# 출처 : https://neurowhai.tistory.com/350
geo_path = './TL_SCCO_CTPRVN.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
geo_str
```

```python
# 데이터프레임과 geojson의 지역명 확인할 열 추가
# 자동화하려 했으나, 갯수가 많지 않아 직접 입력
data_df['지역명'] = ['서울특별시', '경기도', '대구광역시', '인천광역시', '부산광역시', '경상남도', '경상북도', '충청남도', '강원도', '충청북도', '광주광역시', '울산광역시', '대전광역시', '전라북도', '전라남도', '제주도', '세종특별자치시']
data_df.head()
os.getcwd
```

```python
# 인구수, 확진자_합계를 졍수로 변경
data_df['인구수'] = pd.to_numeric(data_df['인구수'].str.replace(',', ''))   
data_df['확진자_합계'] = pd.to_numeric(data_df['확진자_합계'].str.replace(',', ''))

data_df.head()
```

```python
# 확진자 합계 지도 시각화
map = folium.Map(location=[36, 127.51], zoom_start=6, tiles='cartodbpositron') 

map.choropleth(geo_data = geo_str, # geojson 데이터
               data = data_df,  # 시각화 할 데이터
               columns = ('지역명', '확진자_합계'), # geojson과 매칭할 컬럼, 시각화 할 컬럼
               fill_color = 'OrRd', # 색 조힙
               key_on = 'feature.properties.CTP_KOR_NM', # geojson 내부 id값
               legend_name='확진자 현황(명)') # 제목

map
```

```python
map.save(f'{day}_전국_확진자_지도시각화.html')
```

```python
# 인구대비 확진자 컬럼 추가
data_df['인구대비_확진자(%)'] = (data_df['확진자_합계'] / data_df['인구수']) * 100  

data_df.head()
```

```python
# 인구대비 확진자 합계 지도 시각화
map = folium.Map(location=[36, 127.51], zoom_start=6, tiles='cartodbpositron') 

map.choropleth(geo_data = geo_str, 
               data = data_df, 
               columns = ('지역명', '인구대비_확진자(%)'), 
               fill_color = 'OrRd', 
               key_on = 'feature.properties.CTP_KOR_NM',
               legend_name='인구대비_확진자(%)') 

map
```

```python
map.save(f'{day}_전국_인구대비_확진자_지도시각화.html')
```
